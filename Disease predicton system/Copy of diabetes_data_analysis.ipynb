This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

columns:


*pregnant:	Number of times pregnant

*glucose	Plasma glucose concentration (glucose tolerance test)

*pressure	Diastolic blood pressure (mm Hg)

*triceps	Triceps skin fold thickness (mm)

*insulin	2-Hour serum insulin (mu U/ml)

*mass	Body mass index (weight in kg/(height in m)\^2)

*pedigree	Diabetes: pedigree function, this indicates the function which scores likelihood of diabetes based on family history

*age:	Age (years)

*diabetes	Class variable (test for diabetes)

```python
from sklearn import preprocessing
 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'species'.
df['Pregnancies']= label_encoder.fit_transform(df['Pregnancies'])
df['Pregnancies'].unique()
df['Glucose']= label_encoder.fit_transform(df['Glucose'])
df['Glucose'].unique()
df['BloodPressure']= label_encoder.fit_transform(df['BloodPressure'])
df['BloodPressure'].unique()
df['SkinThickness']= label_encoder.fit_transform(df['SkinThickness'])
df['SkinThickness'].unique()
df['Insulin']= label_encoder.fit_transform(df['Insulin'])
df['Insulin'].unique()
df['BMI']= label_encoder.fit_transform(df['BMI'])
df['BMI'].unique()
df['DiabetesPedigreeFunction']= label_encoder.fit_transform(df['DiabetesPedigreeFunction'])
df['DiabetesPedigreeFunction'].unique()
df['Age']= label_encoder.fit_transform(df['Age'])
df['Age'].unique()
df['Outcome']= label_encoder.fit_transform(df['Outcome'])
df['Outcome'].unique()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(["Outcome"], axis=1)
y = df['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train_pca, y_train)
model_rf.score(X_test_pca, y_test)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(["Outcome"], axis=1)
y = df['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python

from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, y_pred)
print('roc_auc_score for DecisionTree: ', roc_auc_score(y_test, y_pred))
plt.subplots(1, figsize=(10,10))
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
