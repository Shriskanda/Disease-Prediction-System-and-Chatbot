This data set contains 416 liver patient records and 167 non liver patient records collected from North East of Andhra Pradesh, India. The "Dataset" column is a class label used to divide groups into liver patient (liver disease) or not (no disease). This data set contains 441 male patient records and 142 female patient records.

Any patient whose age exceeded 89 is listed as being of age "90".

Columns:

Age of the patient
Gender of the patient
Total Bilirubin
Direct Bilirubin
Alkaline Phosphotase
Alamine Aminotransferase
Aspartate Aminotransferase
Total Protiens
Albumin
Albumin and Globulin Ratio
Dataset: field used to split the data into two sets (patient with liver disease, or no disease)

```python
df["Gender"]=[1 if i.strip()=='Male' else 0 for i in df["Gender"]]
df.head()
```

```python
from sklearn import preprocessing
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'species'.
df['Age']= label_encoder.fit_transform(df['Age'])
df['Age'].unique()
df['Gender']= label_encoder.fit_transform(df['Gender'])
df['Gender'].unique()
df['Total_Bilirubin']= label_encoder.fit_transform(df['Total_Bilirubin'])
df['Total_Bilirubin'].unique()
df['Direct_Bilirubin']= label_encoder.fit_transform(df['Direct_Bilirubin'])
df['Direct_Bilirubin'].unique()
df['Alkaline_Phosphotase']= label_encoder.fit_transform(df['Alkaline_Phosphotase'])
df['Alkaline_Phosphotase'].unique()
df['Aspartate_Aminotransferase']= label_encoder.fit_transform(df['Aspartate_Aminotransferase'])
df['Aspartate_Aminotransferase'].unique()
df['Alamine_Aminotransferase']= label_encoder.fit_transform(df['Alamine_Aminotransferase'])
df['Alamine_Aminotransferase'].unique()
df['Total_Protiens']= label_encoder.fit_transform(df['Total_Protiens'])
df['Total_Protiens'].unique()
df['Albumin']= label_encoder.fit_transform(df['Albumin'])
df['Albumin'].unique()
df['Albumin_and_Globulin_Ratio']= label_encoder.fit_transform(df['Albumin_and_Globulin_Ratio'])
df['Albumin_and_Globulin_Ratio'].unique()
df['Dataset']= label_encoder.fit_transform(df['Dataset'])
df['Dataset'].unique()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(["Dataset"], axis=1)
y = df['Dataset']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

Rand_FR = RandomForestRegressor(n_estimators=500, n_jobs=-1)

Rand_FR.fit(X_train, y_train)
y_pred_rf = Rand_FR.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


# score = sklearn.metrics.r2_score(y_test, y_pred_rf, sample_weight=None)
score=r2_score(y_test,y_pred_rf)
print('R2 socre is',score)
print('mean_sqrd_error is =',mean_squared_error(y_test,y_pred_rf))
print('root_mean_squared error of is =',np.sqrt(mean_squared_error(y_test,y_pred_rf)))
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train_pca, y_train)
model_rf.score(X_test_pca, y_test)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, y_pred_rf)
print('roc_auc_score for Random Forest Classifier: ', roc_auc_score(y_test, y_pred_rf))
plt.subplots(1, figsize=(10,10))
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
